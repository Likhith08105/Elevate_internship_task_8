```python
import pandas as pd
```

```python
df = pd.read_csv("Superstore_Sales.csv", parse_dates=['Order Date'], infer_datetime_format=True)
df.head()
```

```python
df = df.dropna(how='all')
df.columns = [c.strip() for c in df.columns]
df.head()
```

```python
for col in ['Sales', 'Profit']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
df.head()
```

```python
if 'Category' in df.columns:
    df['Category'] = df['Category'].fillna('Unknown')
df.head()
```

```python
df['Month-Year'] = df['Order Date'].dt.to_period('M').astype(str)
df['Month Year Label'] = df['Order Date'].dt.strftime('%b %Y')
df.head()
```

```python
df.to_csv("Superstore_Sales_cleaned.csv", index=False)
print("Saved Superstore_Sales_cleaned.csv with", len(df), "rows")
```
